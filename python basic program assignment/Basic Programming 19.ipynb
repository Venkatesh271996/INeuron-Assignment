Question1 Create a function that takes a string and returns a string in which each character is repeated once. 

Examples 

double_char("String") ➞ "SSttrriinngg" 

double_char("Hello World!") ➞ "HHeelllloo WWoorrlldd!!" 

double_char("1234!_ ") ➞ "11223344!!__ "

```python
def double_char(word):
  string_word="".join([i*2 for i in word])
  return string_word

print('double_char("String") ➞ ',double_char("String"))
print('double_char("Hello World!") ➞ ',double_char("Hello World!"))
print('double_char("1234!_ ") ➞ ',double_char("1234!_ "))
```

Question2
Create a function that reverses a boolean value and returns the string &quot;boolean expected&quot;if another variable type is given.

Examples

reverse(True) ➞ False

reverse(False) ➞ True

reverse(0) ➞ &quot;boolean expected&quot;

reverse(None) ➞ &quot;boolean expected&quot;

```python
def reverse(string):
  if type(string)==bool:
    return not string
  else:
    return "boolean expected"


print("reverse(True) ➞", reverse(True))

print("reverse(False) ➞",reverse(False))

print("reverse(0) ➞",reverse(0))

print("reverse(None) ➞",reverse(None))
```

Question3 Create a function that returns the thickness (in meters) of a piece of paper after folding it n number of times. The paper starts off with a thickness of 0.5mm. Examples num_layers(1) ➞ "0.001m"

Paper folded once is 1mm (equal to 0.001m)
num_layers(4) ➞ "0.008m"

Paper folded 4 times is 8mm (equal to 0.008m)
num_layers(21) ➞ "1048.576m"

Paper folded 21 times is 1048576mm (equal to 1048.576m)

```python
def num_layers(input_number):
  thickness=.5
  for i in range(input_number):
    thickness = thickness*2
  return thickness/1000

print(f'num_layers(1) ➞ {num_layers(1)}m')
print(f'num_layers(4) ➞ {num_layers(4)}m')
print(f'num_layers(21) ➞ {num_layers(21)}m')
```

Question4

Create a function that takes a single string as argument and returns an ordered list containing the indices of all capital letters in the string. Examples 

index_of_caps("eDaBiT") ➞ [1, 3, 5] 

index_of_caps("eQuINoX") ➞ [1, 3, 4, 6] 

index_of_caps("determine") ➞ [] 

index_of_caps("STRIKE") ➞ [0, 1, 2, 3, 4, 5] 

index_of_caps("sUn") ➞ [1]

```python
def index_of_caps(input_string):
  cap_list=[input_string.index(ele) for ele in input_string if ele.isupper()]
  return cap_list
print('index_of_caps("eDaBiT") ➞',index_of_caps("eDaBiT"))
print('index_of_caps("eQuINoX") ➞',index_of_caps("eQuINoX"))
print('index_of_caps("determine") ➞',index_of_caps("determine"))
print('index_of_caps("STRIKE") ➞',index_of_caps("STRIKE"))
print('index_of_caps("sUn") ➞',index_of_caps("sUn"))
```

Question5 

Using list comprehensions, create a function that finds all even numbers from 1 to the given number. Examples 

find_even_nums(8) ➞ [2, 4, 6, 8] 

find_even_nums(4) ➞ [2, 4] 

find_even_nums(2) ➞ [2]

```python
def find_even_nums(input_number):
  even_list=[number for number in range(1,input_number+1) if number%2==0]
  return even_list

print(f'find_even_nums(8) ➞',find_even_nums(8))
print(f'find_even_nums(4) ➞',find_even_nums(4))
print(f'find_even_nums(2) ➞',find_even_nums(2))
```

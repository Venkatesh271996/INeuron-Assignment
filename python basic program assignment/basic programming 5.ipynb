1. Write a Python Program to Find LCM?

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

def compute_lcm(x, y):
   if x > y:
       greater = x
   else:
       greater = y
   while(True):
       if((greater % x == 0) and (greater % y == 0)):
           lcm = greater
           break
       greater += 1
   return lcm

print(f"The L.C.M. of {num1},{num2} is", compute_lcm(num1, num2))
```

2. Write a Python Program to Find HCF?

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

def compute_hcf(x, y):
    if x > y:
        smaller = y
    else:
        smaller = x
    for i in range(1, smaller+1):
        if((x % i == 0) and (y % i == 0)):
            hcf = i 
    return hcf

print("The H.C.F. is", compute_hcf(num1, num2))
```

3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
number=int(input("Enter a decimal number: "))
print("The binary number for {number} is",bin(number))
print("The octal number for {number} is",oct(number))
print("The hexadecimal number for {number} is",hex(number))
```

4. Write a Python Program To Find ASCII value of a character?

```python
single_character=input("Enter a single character:")
if len(single_character)==1:
  print(f"The ASCII value of {single_character} is",ord(single_character))
else:
  print("kindly enter a single character")
```

5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
number_1=int(input("Enter the first number: "))
number_2=int(input("Enter the second number: "))

operation=input('''enter the arithmetic operation:
"+"
"-"
"*"
"/" ''')
if operation=="+":
  print(f"{number_1} + {number_2} =",number_1+number_2)

elif operation=="-":
  print(f"{number_1} - {number_2} =",number_1-number_2)

elif operation=="*":
  print(f"{number_1} * {number_2} =",number_1*number_2)

if operation=="/":
  print(f"{number_1} / {number_2} =",number_1/number_2)
```

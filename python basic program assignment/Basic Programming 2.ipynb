1. Write a Python program to convert kilometers to miles?



```python
kilometer=float(input("Enter the kilometers: "))
miles=kilometer*0.621371
print(f"{kilometer} kilometers is equal to {miles} miles")
```

2. Write a Python program to convert Celsius to Fahrenheit?

```python
celcius = float(input("Enter the temperature in degree celsius: "))
fahrenheit=(celcius*9/5)+32
print(f"The {celcius} degree celcius is equal to {fahrenheit} fahrenheit")
```

3. Write a Python program to display calendar?

```python
import calendar

year = int(input("Enter calender year: "))
print(calendar.calendar(year))
```

4. Write a Python program to solve quadratic equation?

```python
import cmath  

a = float(input('Enter a value: '))  
b = float(input('Enter b value: '))  
c = float(input('Enter c value: '))  
  
discriminant = (b**2) - (4*a*c)  
  
sol1 = (-b-cmath.sqrt(discriminant))/(2*a)  
sol2 = (-b+cmath.sqrt(discriminant))/(2*a)  
print('The solution are {0} and {1}'.format(sol1,sol2))  
```

5. Write a Python program to swap two variables without temp variable?

```python
first_number = int(input("Enrter the first number: "))
second_number = int(input("Enter the second number: "))
print("The first and second number before swapping are",first_number,"and",second_number)
first_number, second_number = second_number, first_number
print("The first and second number after swapping are",first_number,"and",second_number)
```

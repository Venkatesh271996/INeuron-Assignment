1. Write a Python Program to Find the Factorial of a Number?

```python
fact=1
number=int(input("Enter the number: "))
for i in range(1,number+1):
  fact=fact*i
print(f"The factorial of the number {number} is {fact}")
```

2. Write a Python Program to Display the multiplication Table?

```python
number=int(input("enter the number for which you want the multiplication table "))
entries=int(input("Enter the number of entries"))
for i in range(1,entries+1):
  print(f"{number}*{i} =",number*i)
```

3. Write a Python Program to Print the Fibonacci sequence?

```python
number=int(input("enter the number of elements you want to display: "))
fib_list=[0,1]
for i in range(2,number):
  fib_list.append(fib_list[i-1]+fib_list[i-2])

for i in fib_list:
  print(i)
```

4. Write a Python Program to Check Armstrong Number?

```python
number=int(input("Enter a number "))
check_number=number
arm_num=0
while number>0:
  a=number%10
  number=number//10
  arm_num=arm_num+a**3
if check_number==arm_num:
  print(f"{check_number} is a armstrong number")
else:
  print(f"{check_number} is not a armstrong number")
```

6. Write a Python Program to Find the Sum of Natural Numbers?

```python
number=int(input("Enter the number: "))
natural_number_sum=number*(number+1)/2
print(f"The sum of all the natural numbers for the {number} is {natural_number_sum}")
```

Question 1 

Create a function that takes a number as an argument and returns True or False depending on whether the number is symmetrical or not. A number is symmetrical when it is the same as its reverse. 

Examples 

is_symmetrical(7227) ➞ True 

is_symmetrical(12567) ➞ False 

is_symmetrical(44444444) ➞ True 

is_symmetrical(9939) ➞ False 

is_symmetrical(1112111) ➞ True

```python
def is_symmetrical(input_number):
  check_number=input_number
  symmetrical_number=0
  while check_number>0:
    reminder=check_number%10
    symmetrical_number=symmetrical_number*10+reminder
    check_number=check_number//10
  
  if input_number==symmetrical_number:
    print(f"is_symmetrical({input_number}) ➞",True)
  else:
    print(f"is_symmetrical({input_number}) ➞",False)

is_symmetrical(7227)
is_symmetrical(12567)
is_symmetrical(44444444)
is_symmetrical(9939)
is_symmetrical(1112111)
```

Question 2 Given a string of numbers separated by a comma and space, return the product of the numbers. Examples multiply_nums("2, 3") ➞ 6 multiply_nums("1, 2, 3, 4") ➞ 24 multiply_nums("54, 75, 453, 0") ➞ 0 multiply_nums("10, -2") ➞ -20

```python
def multiply_nums(input_nums):
  input_list=input_nums.split(",")
  output=1
  for ele in input_list:
    output =output*int(ele)
  return output

print('multiply_nums("2, 3") ➞',multiply_nums("2, 3"))
print('multiply_nums("1, 2, 3, 4") ➞',multiply_nums("1, 2, 3, 4"))
print('multiply_nums("54, 75, 453, 0")  ➞',multiply_nums("54, 75, 453, 0"))
print('multiply_nums("10, -2") ➞',multiply_nums("10, -2"))
```

Question 3 Create a function that squares every digit of a number. 
Examples 

square_digits(9119) ➞ 811181 

square_digits(2483) ➞ 416649 

square_digits(3212) ➞ 9414 

Notes The function receives an integer and must return an integer.

```python
def square_digits(input_number):
  output=""
  string_num_list=list(str(input_number))
  for digit in string_num_list:
    square=str(int(digit)**2)
    output=output+square
  return output

print("square_digits(9119) ➞",square_digits(9119))
print("square_digits(2483) ➞",square_digits(2483))
print("square_digits(3212) ➞",square_digits(3212))
```

Question 4 Create a function that sorts a list and removes all duplicate items from it. Examples 

setify([1, 3, 3, 5, 5]) ➞ [1, 3, 5] 

setify([4, 4, 4, 4]) ➞ [4] 

setify([5, 7, 8, 9, 10, 15]) ➞ [5, 7, 8, 9, 10, 15] 

setify([3, 3, 3, 2, 1]) ➞ [1, 2, 3]

```python
def setify(input_list):
  return list(set(input_list))

print("setify([1, 3, 3, 5, 5]) ➞",setify([1, 3, 3, 5, 5]))
print("setify([4, 4, 4, 4]) ➞",setify([4, 4, 4, 4]))
print("setify([5, 7, 8, 9, 10, 15]) ➞",setify([5, 7, 8, 9, 10, 15]))
print("setify([3, 3, 3, 2, 1]) ➞",setify([3, 3, 3, 2, 1]))
```

Question 5 

Create a function that returns the mean of all digits. 

Examples 

mean(42) ➞ 3 

mean(12345) ➞ 3 

mean(666) ➞ 6
 Notes  The mean of all digits is the sum of digits / how many digits there are (e.g. mean of digits in 512 is (5+1+2)/3(number of digits) = 8/3=2).  The mean will always be an integer.

```python
def mean(input_number):
  sum=0
  no_of_digits=len(str(input_number))
  while input_number>0:
    reminder=input_number%10
    sum = sum + reminder
    input_number=input_number//10
  output=int(sum/no_of_digits)
  return output

print("mean(42) ➞",mean(42))
print("mean(12345) ➞",mean(12345))
print("mean(666) ➞",mean(666))

```

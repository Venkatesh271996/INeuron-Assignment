1. Write a Python Program to find sum of array?

```python
no_of_array_elements=int(input("Enter the number of array elements: "))
array_elements=[]
for i in range(1,no_of_array_elements+1):
  elements=int(input(f"Enter the {i} number: "))
  array_elements.append(elements)
Sum_of_array_elements=sum(array_elements)
print(f"The sum of {array_elements} is",Sum_of_array_elements)
```

2. Write a Python Program to find largest element in an array?

```python
array_elements=eval(input("Enter the array: "))
print(f"The largest element in the array {array_elements} is",max(array_elements))
```

3. Write a Python Program for array rotation?

```python
array_elements=eval(input("Enter the array: "))
print(f"The array rotation of {array_elements} is",array_elements[::-1])
```

4. Write a Python Program to Split the array and add the first part to the end?

```python
array_elements=eval(input("Enter the array: "))
print(f"The sum of first part and the last part of the {array_elements} is",array_elements[0]+array_elements[-1])
```

5. Write a Python Program to check if given array is Monotonic?

```python
array_elements=eval(input("Enter the array: "))
array_unique_element=list(set(array_elements))
if len(array_elements)==len(array_unique_element):
  print(f"The array {array_elements} is Monotonic")
else:
  print(f"The array {array_elements} is not Monotonic")
```

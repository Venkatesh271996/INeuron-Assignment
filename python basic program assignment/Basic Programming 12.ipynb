1. Write a Python program to Extract Unique values dictionary values?

```python
Vinoth={"maths":60,"English":90,"CSC":60,"Social_science":100,"Hindi":70,"Science":90}
print("The unique value in the dictionary is",list(set(Vinoth.values())))
```

2. Write a Python program to find the sum of all items in a dictionary?

```python
Vinoth={"maths":60,"English":90,"CSC":60,"Social_science":100,"Hindi":70,"Science":90}
print("The total marks scored by vinoth is",sum(Vinoth.values()))
```

3. Write a Python program to Merging two Dictionaries?

```python
Vinoth_subject_dict_1={"maths":60,"English":90,"CSC":60}
print("The dictionary 1 is",Vinoth_subject_dict_1)
Vinoth_subject_dict_2={"Social_science":100,"Hindi":70,"Science":90}
print("The dictionary 2 is",Vinoth_subject_dict_2)
Vinoth_subject_dict_1.update(Vinoth_subject_dict_2)
print("The merged dictionary is",Vinoth_subject_dict_1)
```

4. Write a Python program to convert key-values list to flat dictionary?

```python
car=[["name","city"],["engine","4 stroke"],["type of vehicle","Petrol"]]
dictonary={}
for i in car:
  dictonary[i[0]]=i[1]
print(dictonary)
```

5. Write a Python program to insertion at the beginning in OrderedDict?


```python
from collections import OrderedDict
dict_1=OrderedDict({"car":"honda","Mobile":"android","os":"windows"})
dict_2={"company":"TCS"}
dict_1.update(dict_2)
dict_1.move_to_end("car",last=False)
print(dict_1)
```

6. Write a Python program to check order of character in string using OrderedDict()?


```python
from collections import OrderedDict
dict_1={"Mobile":"android","os":"windows","car":"honda"}
final_dict=OrderedDict(dict(sorted(dict_1.items())))
print(final_dict)
```

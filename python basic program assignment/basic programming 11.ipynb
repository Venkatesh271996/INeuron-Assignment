1. Write a Python program to find words which are greater than given length k?

```python
string=input("Enter The string: ")
k=int(input("Enter the K Value: "))
word_list=string.split(" ")
print(f"The words which with length greater than {k} are given below:")
for i in word_list:
  if len(i)>k:
    print(i)
  else:
    pass
```

2. Write a Python program for removing i-th character from a string?

```python
string=input("Enter The string: ")
i=int(input("Enter the character number which you want to remove: "))
character_removal=string[:i-1]+string[i:]
print(f"The string ('{string}') after removing the {i} character is",character_removal)
```

3. Write a Python program to split and join a string?

```python
string=input("Enter The string: ")
split_character=input("Enter the character based upon which you want to split: ")
words=string.split(split_character)
print("The words after spliting the string is",words)
joining_character=input("Enter the character based upon which you want to join: ")
print("The Joined string is:",joining_character.join(words))
```

4. Write a Python to check if a given string is binary string or not?

```python
string=input("Enter The string: ")
status=[]
for i in string:
  if i in ["0","1"]:
    status.append(1)
  else:
    status.append(0)

if 0 in status:
  print("The given string is not a binary string")
else:
  print("The given string is binary string")
```

5. Write a Python program to find uncommon words from two Strings?

```python
string_1=input("Enter the first String: ")
string_2=input("Enter the second String: ")
list_1=string_1.split(" ")
list_2=string_2.split(" ")
uncommon_words=[]
for i in list_1:
  if i not in list_2:
    uncommon_words.append(i)
for j in list_2:
  if j not in list_1:
    uncommon_words.append(j)
print(uncommon_words)
```

6. Write a Python to find all duplicate characters in string?

```python
string_1=input("Enter the string: ")
unduplicate_list=[]
duplicate_list=[]
for i in string_1:
  if i not in  unduplicate_list:
    unduplicate_list.append(i)
  else:
    duplicate_list.append(i)
print("The duplicate characters in the string are",list(set(duplicate_list)))
```

7. Write a Python Program to check if a string contains any special character?

```python
string=input("Enter the string: ")
special_characters = "[@_!#$%^&*()<>?/\|}{~:]"
for i in string:
  if i in special_characters:
    result="The string contains special character"
    break
  else:
    result="The string does not contain any special character"
print(result)
```

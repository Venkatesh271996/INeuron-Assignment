1. Write a Python Program to Add Two Matrices?

```python
print("Enter the 1 matrix details")
m=int(input("Enter the no of rows: "))
n=int(input("Enter the no of columns: "))
def create_matrix(y,z):
  row=[]
  for i in range(0,y):
    column=[]
    for j in range(0,z):
      a=int(input(f"Enter the {i} row and {j} column element: "))
      column.append(a)
    row.append(column)
  return row
matrix_1=create_matrix(m,n)
print("Enter the 2nd matrix details")
matrix_2=create_matrix(m,n)
result_row=[]
for i in range(0,m):
  result_column=[]
  for j in range(0,n):
    result_column.append(0)
  result_row.append(result_column)
result=result_row
for i in range(m):
  for j in range(n):
    result[i][j]=matrix_1[i][j]+matrix_2[i][j]
print(result)
```

2. Write a Python Program to Multiply Two Matrices?

```python
print("Enter the 1 matrix details")
m=int(input("Enter the no of rows: "))
n=int(input("Enter the no of columns: "))

def create_matrix(y,z):
  row=[]
  for i in range(0,y):
    column=[]
    for j in range(0,z):
      a=int(input(f"Enter the {i} row and {j} column element: "))
      column.append(a)
    row.append(column)
  return row
matrix_1=create_matrix(m,n)
print("Enter the 2nd matrix details")
matrix_2=create_matrix(m,n)
result_row=[]
for i in range(0,m):
  result_column=[]
  for j in range(0,n):
    result_column.append(0)
  result_row.append(result_column)
result=result_row

for m in range(len(matrix_1)):    
   for n in range(len(matrix_2[0])):    
          for o in range(len(matrix_2)):    
               result[m][n] += matrix_1[m][o] * matrix_2[o][n] # Storing multiplication result in empty matrix  
# Printing multiplication result in the output  
print("The multiplication result of matrix A and B is: ")  
for res in result:    
   print(res)  
```

3. Write a Python Program to Transpose a Matrix?

```python
print("Enter the 1 matrix details")
m=int(input("Enter the no of rows: "))
n=int(input("Enter the no of columns: "))

def create_matrix(y,z):
  row=[]
  for i in range(0,y):
    column=[]
    for j in range(0,z):
      a=int(input(f"Enter the {i} row and {j} column element: "))
      column.append(a)
    row.append(column)
  return row
matrix_1=create_matrix(m,n)

result = [[matrix_1[j][i] for j in range(len(matrix_1))] for i in range(len(matrix_1[0]))]

for r in result:
   print(r)
```

4. Write a Python Program to Sort Words in Alphabetic Order?

```python
number=int(input("Enter the no of words: "))
word_list=[]
for i in range(1,number+1):
  input_word=input(f"enter the {i} word: ")
  word_list.append(input_word)
print("The word list before sorted",word_list)
print("The word list after sorting",sorted(word_list,reverse=False))
```

5. Write a Python Program to Remove Punctuation From a String?

```python
string=input("Enter the string: ")
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
sentence=""
for i in string:
  if i not in punctuations:
    sentence=sentence+i
print(sentence)
```

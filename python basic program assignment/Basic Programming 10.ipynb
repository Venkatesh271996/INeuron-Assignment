1.  Write a Python program to find sum of elements in list?

```python
def sum_element(n):
  list_1=[]
  for i in range(1,n+1):
    elements=int(input(f"Enter the {i} element: "))
    list_1.append(elements)
  elements_sum=sum(list_1)
  return elements_sum

result=int(input("Enter the no of elements: "))
print("The sum of the elements in the list is",sum_element(result))
```

2. Write a Python program to Multiply all numbers in the list?

```python
def sum_element(n):
  list_1=[]
  for i in range(1,n+1):
    elements=int(input(f"Enter the {i} element: "))
    list_1.append(elements)
  list_multiplication=1
  for i in list_1:
    list_multiplication*=i
  return list_multiplication

result=int(input("Enter the no of elements: "))
print("The multiplication of the elements in the list is",sum_element(result))
```

3. Write a Python program to find smallest number in a list?

5. Write a Python program to find second largest number in a list?

```python
def element_sort(n):
  list_1=[]
  for i in range(1,n+1):
    elements=int(input(f"Enter the {i} element: "))
    list_1.append(elements)
  return sorted(list_1,reverse=False)

result=int(input("Enter the no of elements: "))
sorted_element=element_sort(result)
print("The second largest number in the list is",sorted_element[-2])
```

6. Write a Python program to find N largest elements from a list?

```python
def element_sort(n):
  list_1=[]
  for i in range(1,n+1):
    elements=int(input(f"Enter the {i} element: "))
    list_1.append(elements)
  return sorted(list_1,reverse=True)

result=int(input("Enter the no of elements: "))
sorted_element=element_sort(result)
N=int(input("Enter the no of largest elements you want to find:"))
print(f"The {N} largest elements in the list are",sorted_element[0:N])
```

7. Write a Python program to print even numbers in a list?


```python
def list_creation(n):
  list_1=[]
  for i in range(1,n+1):
    elements=int(input(f"Enter the {i} element: "))
    list_1.append(elements)
  return list_1

result=int(input("Enter the no of elements: "))
list_2=list_creation(result)
print("The even numbers in the list are given below:")
for i in list_2:
  if i%2==0:
    print(i)
  else:
    pass
```

10. Write a Python program to Cloning or Copying a list?

```python
def list_cloning(list_elements):
  list_2=list_elements.copy()
  return list_2

list_1=eval(input("Enter the list"))
print(f"The copy of the list {list_1} is",list_cloning(list_1))
```

11. Write a Python program to Count occurrences of an element in a list?

```python
list_1=eval(input("Enter the list"))
element=int(input("Enter the element you want to find:"))
count=0
if element in list_1:
  for elem in list_1:
    if element==elem:
      count=count+1
  print(f"The {element} is occuring {count} times in the {list_1}")
else:
  print(f"The {element} is not in the {list_1}")
```
